```python
import numpy 

def bounding_box(iterable):
    min_x, min_y = np.min(iterable[0], axis=0)
    max_x, max_y = np.max(iterable[0], axis=0)
    return np.array([(min_x, min_y), (max_x, min_y), (max_x, max_y), (min_x, max_y)])
```

```python
def old_bounding_box(coords): # not using numpy

  min_x = 100000 # start with something much higher than expected min
  min_y = 100000
  max_x = -100000 # start with something much lower than expected max
  max_y = -100000

  for item in coords:
    if item[0] < min_x:
      min_x = item[0]

    if item[0] > max_x:
      max_x = item[0]

    if item[1] < min_y:
      min_y = item[1]

    if item[1] > max_y:
      max_y = item[1]

  return [(min_x,min_y),(max_x,min_y),(max_x,max_y),(min_x,max_y)]
```

```python
'''
Read CSV from File 
=LEFT($B2,FIND(" ", $B2))
=RIGHT($B2,LEN($B2)-FIND(" ", $B2))
'''
from numpy import genfromtxt
#def readCSVfile(filename):
filename = 'testclusterdata/DRC Control 1- 2016-05-01_Conv.csv'
print "Reading CSV File"
X = genfromtxt(filename, 
               delimiter=',', skip_header=1)

   
print("Read CSV File to X:", X)
```

```python
##############################################################################
# Compute DBSCAN
from sklearn.cluster import DBSCAN
from sklearn import metrics
#from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler

print ("Started...")
db = DBSCAN(eps=0.012, min_samples=2).fit(X)
core_samples = db.core_sample_indices_
labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
unique_clusters = set(labels)

n_clusters_ = len(unique_clusters) - (1 if -1 in labels else 0)
print('Estimated number of clusters: %d' %n_clusters_)

print ("unclustered",  unique_clusters)
if -1 in labels:
    #rint (unique_clusters[-1] )
    print ("unclustered points: %d." %labels[-1])

print("Silhouette Coefficient: %0.3f"
      % metrics.silhouette_score(X, labels))

###############################################################################
```

```python
##############################################################################
# Plot result
from scipy.spatial import ConvexHull
import pylab as pl
#for p in a:
#    pl.plot(p["long"], p["lat"])

print "Plotting result"
# Black removed and is used for noise instead.
unique_labels = set(labels)
colors = pl.cm.Spectral(numpy.linspace(0, 1, len(unique_labels)))
testing = zip(unique_labels, colors)

#print ("zip(unique_labels, colors)", testing)


for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = 'k'
        markersize = 1
    class_members = [index[0] for index in np.argwhere(labels == k)]
    cluster_core_samples = [index for index in core_samples
                            if labels[index] == k]
    for index in class_members:
        x = X[index]
        
        if index in core_samples and k != -1:
            markersize = 4
        else:
            markersize = 2
        pl.plot(x[1], x[0], 'o', markerfacecolor=col,
                markeredgecolor='k', markersize=markersize)
       
msg = '%s : clusters: %d' % (filename, n_clusters_)
print(msg)
pl.title(msg)
pl.show()
```
